```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
job_titles = []
company_names = []
Locations = []
skills = []
Links = []
```

```python
result = requests.get('https://wuzzuf.net/search/jobs/?q=data%20analysist&a=hpb')
```

```python
page_content = result.content
print(page_content)
```

```python
soup = BeautifulSoup(page_content,'lxml')
print(soup)
```

```python
job_title = soup.find_all('h2', {'class':'css-m604qf'})
print(job_title)
```

```python
company_name = soup.find_all('a',{'class':'css-17s97q8'})
print(company_name)
```

```python
Location = soup.find_all('span',{'class':'css-5wys0k'})
print(Location)
```

```python
job_skills = soup.find_all('div',{'class':'css-y4udm8'})
print(job_skills)
```

```python
for i in range (len(job_title)):
    job_titles.append(job_title[i].text)
    company_names.append(company_name[i].text)
    Locations.append(Location[i].text)
    skills.append(job_skills[i].text)
    Links.append(job_title[i].find('a').attrs['href'])
```

```python
file_list = [job_titles,company_names,skills,Locations,Links]
export = zip_longest(*file_list)
```

```python
with open ("C:/Users/MSI/Desktop/project/python/Course/job_test.csv","w")as MyFile:
    wr = csv.writer(MyFile)
    wr.writerow(["job_title","company_name","skills","Locations","Job_Link"])
    wr.writerows(export)
```
